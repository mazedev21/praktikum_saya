```python
def LinearSearch(arr, N, x):
  for i in range(0, N):
    if arr[i] == x:
      return i
  return "TidakAda"

#Driver code
if __name__ == "__main__":
  arr = [10,30,15,26,56,34,44]
  x = 26
  N = len(arr)

  #function call
  result = LinearSearch(arr, N, x)
  if result == "TidakAda":
    print('Elemen yang dicari tidak ada dalam array')
  else:
    print("Elemen yang dicari berada pada indeks ke-", result)
```

```python
def BinarySearch(arr, l, r, x):
  while l <= r:
    mid = l + (r - 1) // 2

    #Check if x is present at mid
    if arr[mid] == x:
      return mid

    #if x is greater, ignore left half
    elif arr[mid] < x:
      l = mid + 1

    #if x is smaller, ignore right half
    else:
      r = mid - 1
  return "Tidak Ada"

#Driver code
if __name__ == "__main__":
  arr = [10,30,15,26,56,34,44]
  x = 34

  #function call
  result = BinarySearch(arr, 0, len(arr)-1, x)
  if result == "Tidak Ada":
    print('Maaf elemen yang dicari tidak ada dalam array')
  else:
    print('Elemen berada di indeks ke-', result)
```

```python
def BinarySearch(arr, l, r, x):
  while r >= l:
    mid = l + (r - 1) // 2

    #Check if x is present at mid
    if arr[mid] == x:
      return mid

    #if x is greater, ignore left half
    elif arr[mid] < x:
      return BinarySearch(arr, mid+1, r, x)

    #if x is smaller, ignore right half
    else:
      return BinarySearch(arr, l, mid-1, x)
  return "Tidak Ada"

#Driver code
if __name__ == "__main__":
  arr = [10,30,15,26,56,34,44]
  x = 34

  #function call
  result = BinarySearch(arr, 0, len(arr)-1, x)
  if result == "Tidak Ada":
    print('Maaf elemen yang dicari tidak ada dalam array')
  else:
    print('Elemen berada di indeks ke-', result)
```

```python
import math

def jumpSearch(arr, x, n):
  #Find block size to be jumped
  step = math.sqrt(n)

  #Finding the block where element is
  #present (if it is present)
  prev = 0
  while arr[int(min(step, n)-1)] < x:
    prev = step
    step += math.sqrt(n)
    if prev >= n:
      return "Tidak Ada"

    #Doing a linear search for x in
    # block beginning with prev
    while arr[int(prev)] < x:
      prev += 1

      #if we reached next block or end
      # of arrat, element is not present
      if prev == min(step, n):
        return "Tidak Ada"

    #If element is found
    if arr[int(prev)] == x:
      return prev

    return "Tidak Ada"

#Driver code
if __name__ == "__main__":
  arr = [0, 1, 1, 2, 3, 4, 5, 8, 13, 21, 34, 55, 89, 144,
         233, 377, 610]
  x = 55
  n = len(arr)

  #Call function
  index = jumpSearch(arr, x, n)

  #Print the index where 'x' is located
  print('Number', x, 'is at index', '%.0f'%index)
```

```python
#Program untuk mencari nilai terendah dan tertinggi

class pair:
  def __init__(self):
    self.min = 0
    self.max = 0

def getMinMax(arr: list, n: int) -> pair:
  minmax = pair()

  #Jika hanya ada satu elemen
  if n == 1:
    minmax.max = arr[0]
    minmax.min = arr[0]
    return minmax

  #Jika hanya ada lebih dari satu elemen
  if arr[0] > arr[1]:
    minmax.max = arr[0]
    minmax.min = arr[1]
  else:
    minmax.max = arr[1]
    minmax.min = arr[0]

  for i in range(2,n):
    if arr[i] > minmax.max:
      minmax.max = arr[i]
    elif arr[i] < minmax.min:
      minmax.min = arr[i]

  return minmax

#Driver code
if __name__ == "__main__":
  arr = [
    85, 78, 92, 88, 76, 81, 90, 87, 73, 84,
    79, 91, 86, 77, 82, 89, 75, 80, 93, 88,
    74, 83, 85, 78, 90, 87, 76, 81, 92, 84,
    79, 88, 86, 77, 82, 91, 75, 80, 89, 83
  ]
  arr_size = 40
  minmax = getMinMax(arr, arr_size)
  print("NIlai minimum adalah:", minmax.min)
  print("Nilai maximum adalah:", minmax.max)
```

```python
def printTwoElements(arr):
  n = len(arr)
  temp = [0] * n
  repeatingNumber = -1
  missingNumber = -1

  for i in range(n):
    temp[arr[i] - 1] += 1
    if temp[arr[i] - 1] > 1:
      repeatingNumber = arr[i]
  for i in range(n):
    if temp[i] == 0:
      missingNumber = i + 1
      break

  print('The repeating number is:', repeatingNumber)
  print('The missing number is', missingNumber)

arr = [7, 3, 4, 5, 5, 6, 2]
printTwoElements(arr)
```

```python
class pair:
  def __init__(self):
    self.num1 = None
    self.num2 = None

def cariDuaNilai(arr: list, x: int) -> pair:
  angka = pair()
  size = len(arr)
  ada = False
  if size > 1:
    for i in range(0, size):
      current = i
      angka.num1 = arr[current]
      for j in range(current+1, size):
        angka.num2 = arr[j]
        if angka.num1 + angka.num2 == x:
          ada = True
          break
      if ada == True:
        print("Apakah ada sepasang elemen yang jumlahnya X:", ada)
        print(f'Terdapat pasangan ({angka.num1}, {angka.num2}) dengan jumlah {x}')
        return
      elif current == size-1 and angka.num1 + angka.num2 != x:
        print('Maaf pasangan angka tidak ada yang cocok')
        return
  else:
    print('Maaf array hanya memiliki satu angka')
    return

arr = [11, 15, 6, 8, 9, 10]
x = 16
cariDuaNilai(arr, x)
```
