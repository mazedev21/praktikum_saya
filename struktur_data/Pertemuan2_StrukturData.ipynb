```python
#Single Linked List

class Node:
  def __init__(self, dataVal):
    self.dataVal = dataVal #menyimpan dataVal ke atribut objek
    self.nextVal = None #referensi ke node berikutnya diset none dulu

class LinkedList:
  def __init__(self):
    self.headVal = None #node pertama default di set None dulu

  #Print Linked List
  def listprint(self):
    current = self.headVal #set pointer ke node pertama
    while current.nextVal: #Perulangan selama current.nextVal tidak bernilai none
      print(current.dataVal, end=" -> ")
      current = current.nextVal #set pointer ke node berikutnya
    if current.nextVal is None: #ketika value selanjutnya dari current = None
      print(current.dataVal, end="\n")

  #Membuat node baru di awal
  def AddBeginning(self,newData):
    NewNode = Node(newData) #membuat objek baru dengan class Node
    NewNode.nextVal = self.headVal #node baru dikoneksikan ke head lama
    self.headVal = NewNode #pindahkan head ke node baru

  #Menambah node di tengah
  def AddInBetween(self,mid_node,newData):
    if mid_node is None: #jika node yang dituju None
      print("The mentioned node is absent")
      return
    NewNode = Node(newData) #membuat objek baru dengan class Node
    NewNode.nextVal = mid_node.nextVal #node baru dikoneksikan dengan NextValue mid_node
    mid_node.nextVal = NewNode #mid_node dikoneksikan dengan node baru

  #Menambah node di akhir
  def AddInLast(self,newData):
    NewNode = Node(newData) #membuat objek baru dengan class Node
    if self.headVal is None: #jika headValue None
      self.headVal = NewNode #set node baru menjadi head
      return
    last = self.headVal
    while(last.nextVal): #selama last.nextVal bukan None
      last = last.nextVal
      #loop berhenti ketika last.nextVal == None
    last.nextVal = NewNode #jadikan node baru menjadi node terakhhir

  #hapus node dengan key
  def RemoveNode(self, Removekey):
    current = self.headVal #set pointer ke head
    if current is not None: #jika pointer saat ini tidak None
      if current.dataVal == Removekey: #jika pointer cocok dengan key
        self.headVal = current.nextVal #set pointer nexVal menjadi head
        current = None #set pointer saat ini menjadi None
        return
    while current is not None: #Selama pointer tidak None
      if current.dataVal == Removekey: #Jika pointer saat ini cocok dengan key
        break #hentikan loop
      prev = current #masukkan nilai pointer saat ini ke var prev
      current = current.nextVal #ubah pointer ke value selanjutnya
    if current == None: #Jika pointer saat ini None
      return print("Key tidak ditemukan")
    prev.nextVal = current.nextVal #set nilai setelah pointer menjadi nextValue milik prev
    current = None #set pointer saat ini dengan None

  #hapus node pertama
  def RemoveFirst(self):
    head = self.headVal #masukan nilai head ke var head
    self.headVal = head.nextVal #set nilai selanjutnya dari head menjadi headVal
    head = None #hapus var nilai head sebelumnya

  #hapus node terakhir
  def RemoveEnd(self):
    current = self.headVal #set pointer ke head
    if current == None: #untuk menangani apabila tak ada data dalam list
      return print('Maaf tidak ada data tersedia')
    if current.nextVal is None: #Untuk menangani apabila hanya terdapat 1 node
      self.headVal = None
      return
    while current.nextVal is not None: #perulangan ketika current bukan None
      prev = current #masukkan objek current ke prev
      current = current.nextVal #ubah pointer ke node berikutnya
    prev.nextVal = None #Hapus nilai selanjutnya dari prev

#Create Nodes
node1 = Node("Januari")
node2 = Node("Februari")
node3 = Node("Maret")
node4 = Node("April")


Li = LinkedList()
Li.headVal = node1 #set node1 to headValue

#Link First Node to Second Node
Li.headVal.nextVal = node2

#Link Second Node to Third Node
node2.nextVal = node3

#Link Third Node to Fourth Node
node3.nextVal = node4

print(node2.dataVal)

print("\nTampilkan linked list")
Li.listprint()

print("\nSetelah ditambah 'Start' diawal")
Li.AddBeginning("Start") #Menambahkan elemen di awal
Li.listprint() #tampilkan linked list

print("\nSetelah ditambah 'Mei' setelah node4")
Li.AddInBetween(node4,"Mei") #menambahkan elemen setelah node4
Li.listprint() #tampilkan linked list

print("\nSetelah ditambah 'Selesai' diakhir")
Li.AddInLast("Selesai") #menambahkan "Selesai" di akhir list
Li.listprint() #tampilkan linked list

print("\nMenghapus node dengan key = 'Maret'")
Li.RemoveNode("Selesai") #menghapus node dengan key = "Maret"
Li.listprint() #tampilkan linked list

print("\nMenghapus node pertama")
Li.RemoveFirst() #Menghapus node pertama
Li.listprint() #tampilkan linked list

print("\nMenghapus node akhir")
Li.RemoveEnd() #menghapus node terakhir
Li.listprint() #tampilkan linked list
```

```python
#DOUBLE LINKED LIST

class Node:
  def __init__(self, dataVal):
    self.dataVal = dataVal #menyimpan dataVal ke atribut objek
    self.nextVal = None #referensi ke node berikutnya diset none dulu
    self.prevVal = None #referensi ke node sebelumnya diset none dulu

class LinkedList:
  def __init__(self):
    self.headVal = None #node pertama default di set None dulu

  def listprint(self):
    current = self.headVal #set pointer ke node pertama
    while current.nextVal: #Perulangan selama current.nextVal tidak bernilai none
      print(current.dataVal, end=" <=> ")
      current = current.nextVal #set pointer ke node berikutnya
    if current.nextVal is None: #ketika value selanjutnya dari current = None
      print(current.dataVal, end="\n")

 #Menambah node di tengah
  def AddInBetween(self,mid_node,newData):
    if mid_node is None: #jika node yang dituju None
      print("The mentioned node is absent")
      return
    NewNode = Node(newData) #membuat objek baru dengan class Node
    NewNode.nextVal = mid_node.nextVal #node baru dikoneksikan dengan NextValue mid_node
    NewNode.prevVal = mid_node #NewNode prevValue dikoneksikan ke mid_node
    NewNode.nextVal.prevVal = NewNode #koneksikan  node baru dengan node selanjutnya
    mid_node.nextVal = NewNode #mid_node dikoneksikan dengan node baru

  #hapus node dengan key
  def RemoveNode(self, Removekey):
    current = self.headVal #set pointer ke head
    if current is None: #ketika pinter head saat ini None
      return print("Maaf tidak ada data tersedia")

    #Jika head yang dihapus
    if current.dataVal == Removekey:
      self.headVal = current.nextVal #Ubah node selanjutnya mennjadi head
      if self.headVal is not None:
        self.headVal.prevVal = None #Menghilangkan prevVal dari head yang baru
      return

    #Cari node
    while current: #ketika current tidak None
      if current.dataVal == Removekey: #ketika key cocok dengan value, hentikan loop
        break
      prev = current #masukkan object current ke var prev
      current = current.nextVal #ubah pointer ke node selanjutnya

    #Key tidak ditemukan
    if current is None:
      return print("Maaf key tidak ditemukan")

    prev.nextVal = current.nextVal #next value dari pointer saat ini di set jadi next value prev
    if prev.nextVal is not None: #ketika next value dari prv bukan None
      prev.nextVal.prevVal = prev #maka set preValue dari node setelahnya dengan prev
    current = None #hapus current

#Create Nodes
node1 = Node("Semarang")
node2 = Node("Magelang")
node3 = Node("Jogja")

Li = LinkedList()
Li.headVal = node1 #set node 1 menjadi node pertama

#Link node1 dengan node2
Li.headVal.nextVal = node2
node2.prevVal = Li.headVal

#Link node2 dengan node3
node2.nextVal = node3
node3.prevVal = node2

print("Double Linked List Rute Semarang-Jogja")
Li.listprint()

print("\nSetelah ditambahkan Temanggung")
Li.AddInBetween(node1,"Temanggung")
Li.listprint()

print("\nSetelah Temanggung dihapus")
Li.RemoveNode("Temanggung")
Li.listprint()
```

```python
#Circular Linked List

class Node:
  def __init__(self, dataVal):
    self.dataVal = dataVal #menyimpan dataVal ke atribut objek
    self.nextVal = None #referensi ke node berikutnya diset none dulu

class LinkedList:
  def __init__(self):
    self.headVal = None #node pertama default di set None dulu

  def listprint(self):
    current = self.headVal #set pointer ke node pertama
    if current is None:
      return print("Maaf data tidak tersedia")
    while current: #Perulangan selama current tidak bernilai none
      print(current.dataVal, end=" -> ")
      current = current.nextVal #set pointer ke node berikutnya
      if current == self.headVal:
        break
    print(current.dataVal, end="\n")

  #Menambah node di akhir
  def AddInLast(self,newData):
    NewNode = Node(newData) #membuat objek baru dengan class Node
    if self.headVal is None: #jika headValue None
      self.headVal = NewNode #set node baru menjadi head
      return
    current = self.headVal
    while current.nextVal is not self.headVal: #selama last.nextVal bukan None
      current = current.nextVal
      #loop berhenti ketika last.nextVal == None
    current.nextVal = NewNode #jadikan node baru menjadi node terakhhir
    NewNode.nextVal = self.headVal #koneksikan node baru dengan head

  #hapus node terakhir
  def RemoveEnd(self):
    current = self.headVal #pointer di set ke node pertama
    if current == None: #jika pointer saat ini none
      return print("Maaf tidak ada nilai yang tersedia")

    #Jika hanya ada satu node
    elif current.nextVal == self.headVal:
      current = None #hapus nilai pertama
      return

    #Jika lebih dari satu node
    while current.nextVal is not self.headVal: #Selama current nextValue bukan nilai pertama
      prev = current #simpan objek saat ini ke prev
      current = current.nextVal #ubah pointer ke node selanjutnya

    #Ketika sudah sampai node terakhir
    prev.nextVal = current.nextVal #nextValue dari objek sebelumnya dikoneksikan dengan node pertama
    current = None #node saat ini dihapus

#Create Nodes
node1 = Node("Masuk Goa")
node2 = Node("Menang")
node3 = Node("Sombong")

Li = LinkedList()
Li.headVal = node1

# Link node1 dengan node2
Li.headVal.nextVal = node2

# Link node2 dengan node3
node2.nextVal = node3

# Link node3 dengan head
node3.nextVal = Li.headVal

print("Siklus Fans MU")
Li.listprint()

print("\nSetelah 'Kalah' ditambahkan")
Li.AddInLast("Kalah")
Li.listprint()

print("\nSetelah node terakhir dihapus")
Li.RemoveEnd()
Li.listprint()
```
